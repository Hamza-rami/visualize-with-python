## Bar sharts

```python
import matplotlib.pyplot as plt
import numpy as np


categories = np.array(["Grains", "Fruit", "Vegatables", "protien", "Dairy", "sweets"])
Values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, Values, color="skyblue")
plt.title("daily consumption")
plt.xlabel("Food", fontsize=15, color="lightgrey")
plt.ylabel("Quentity", fontsize=15, color="lightgrey")
```

## pie sharts

```python
categories = ["Freshman", "Sophomors", "Juniors", "Seniors"]
Values = np.array([400, 300, 200, 100])
colors = ["grey", "lightyellow", "skyblue", "lightgreen"]
plt.pie(Values, labels=categories, autopct="%1.0f%%", colors=colors, explode=[0, 0, 0, 0.2], shadow=True)
plt.title("Hrami code college")
```

## Scatter graphs

```python
hour = np.array([2, 3, 5, 8, 12, 16]) # Hours studied
grad = np.array([8, 10, 13, 15, 18, 19]) # Grades


hour1 = np.array([2, 3, 5, 8, 12, 16]) # Hours studied
grad1 = np.array([10, 12, 15, 17, 19, 20]) # Grades

plt.scatter(hour, grad, color="grey", alpha=1, s=50, label="class A")
plt.scatter(hour1, grad1, color="skyblue", alpha=1, s=50, label="class B")
plt.legend()
plt.title("Test Scores")
plt.xlabel("Hours studied")
plt.ylabel("Grades")
```

## Histogram

```python
s = np.random.normal(loc=10, scale=5, size=100)
s = np.clip(s, 0, 20)
plt.hist(s, color="lightblue", edgecolor="black")
plt.title("Exam Scores")
plt.xlabel("scores")
plt.ylabel("number of student")
```

## subplots

```python
x = np.array([1,2,3,4,5])

fig, ax = plt.subplots(2, 2)
ax[0,0].plot(x, x*2, color="red")
ax[0,0].set_title("x*2")
ax[0,1].plot(x, x**2, color="Blue")
ax[0,1].set_title("x**2")
ax[1,0].plot(x, x**3, color="yellow")
ax[1,0].set_title("x**3")
ax[1,1].plot(x, x**4, color="green")
ax[1,1].set_title("x**4")
plt.tight_layout()
```
