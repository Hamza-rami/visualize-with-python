## Plot customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2000, 2005, 2010, 2015, 2020, 2025])
y1= np.array([10, 40, 90, 100, 70, 200])
y2 = np.array([300, 280, 200, 310, 20, 50])
y3 = np.array([100, 120, 160, 200, 200, 300])
line_style = dict(marker="o",
                    markersize="7",
                    markerfacecolor="red",
                    markeredgecolor="red",
                    linestyle="-",
                    linewidth=3)
plt.plot(x, y1, color="orange", **line_style)
plt.plot(x, y2, color="cyan", **line_style)
plt.plot(x, y3, color="blue", **line_style)
plt.xlabel("Year From 2000 to 2025")
plt.ylabel("Number of sales")
plt.title("class size", fontsize="15",
                        color="grey")
plt.tick_params(axis='both', colors="black")
plt.grid()
plt.show()
```
