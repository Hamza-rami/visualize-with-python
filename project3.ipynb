## pandas + matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data.csv")
type1_count = df["Type1"].value_counts(ascending=True)
plt.barh(type1_count.index, type1_count.values, color="lightgreen", edgecolor="black")
plt.title("# of pokemon by primary type")
plt.xlabel("count")
plt.ylabel("Type")
```
